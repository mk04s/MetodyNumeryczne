```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return x**3 - 5*x + 7


x1 = np.linspace(-1, 1, 400) 
y1 = f(x1)


x2 = np.linspace(-10, 20, 400)
y2 = f(x2)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))


ax1.plot(x1, y1, color='blue', linewidth=2, label=r'$f(x) = x^3 - 5x + 7$')
ax1.set_title(r'Wykres w przedziale $x \in \langle-1, 1\rangle$', fontsize=14)
ax1.set_xlabel('Oś X', fontsize=12)
ax1.set_ylabel('Oś Y', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(loc='best')
ax1.axhline(0, color='black', linewidth=1.2)
ax1.axvline(0, color='black', linewidth=1.2)


ax2.plot(x2, y2, color='red', linewidth=2, label=r'$f(x) = x^3 - 5x + 7$')
ax2.set_title(r'Wykres w przedziale $x \in \langle-10, 20\rangle$', fontsize=14)
ax2.set_xlabel('Oś X', fontsize=12)
ax2.set_ylabel('Oś Y', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend(loc='best')
ax2.axhline(0, color='black', linewidth=1.2)
ax2.axvline(0, color='black', linewidth=1.2)


fig.suptitle(r'Wizualizacja funkcji $f(x) = x^3 - 5x + 7$', fontsize=16, fontweight='bold')

plt.tight_layout()


plt.show() 
```

 Porównywanie wizualne i statystyczne.
 Gdy macierze są ogromne, nie będziesz przeglądać tysięcy liczb. Wtedy warto ocenić ich "bliskość" globalnie:Macierz różnic (Heatmapa): Odejmujesz jedną macierz od drugiej ($C = A - B$) i wizualizujesz wynik. Jeśli macierze są podobne, powinieneś zobaczyć głównie kolor odpowiadający zeru.Norma macierzowa (np. Frobeniusa): Obliczenie jednej liczby, która mówi, jak bardzo macierze się różnią.W NumPy: np.linalg.norm(A - B). Im wynik bliższy zeru, tym macierze są bardziej identyczne.

```python
import numpy as np

A = np.array([[1.0, 2.0], [3.0, 4.0]])
B = np.array([[1.0, 2.0], [3.0, 4.000000001]])


print(np.array_equal(A, B))  


print(np.allclose(A, B, atol=1e-8))  


roznica = np.linalg.norm(A - B)
print(f"Błąd całkowity: {roznica}")
```

```python
import numpy as np
from scipy.linalg import qr

A = np.random.rand(10, 10)

Q, R = qr(A)


Q_T = Q.T
Q_inv = np.linalg.inv(Q)


print("Q:\n", Q)
print("\nR:\n", R)


print("\nCzy Q^T = ", np.allclose(Q_T, Q_inv))
```
