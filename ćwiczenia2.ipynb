Zadanie 1

```python
import math


e_exact = math.e

def approx_e(n_max):
    return sum(1 / math.factorial(n) for n in range(n_max + 1))

n_values = [5, 10, 20]

for n in n_values:
    e_approx = approx_e(n)
    abs_error = abs(e_exact - e_approx)
    rel_error = abs_error / abs(e_exact)
    print(f"Dla n = {n}:")
    print(f"  Przybliżenie: {e_approx:.16f}")
    print(f"  Błąd bezwzględny: {abs_error:.4e}")
    print(f"  Błąd względny: {rel_error:.4e}\n")
```

Zadanie 2c: Liczby dokładnie reprezentowane

```python
import math

print("Liczby dokładnie reprezentowane w [1, 2):")

print("Liczba dokładnie reprezentowanych liczb w [1,2): 2^52 =", 252)

print("\nReprezentacja liczby 1.0")
print("1.0 ->", (1.0).hex())
print("Wniosek: mantysa składa się z samych zer")

print("\nNajmniejsza liczba > 1:")
next_after_1 = math.nextafter(1.0, 2.0)
print("Najmniejsza liczba większa od 1:", next_after_1)
print("hex:", next_after_1.hex())

diff = next_after_1 - 1.0
print("Różnica (czyli dodana wartość):", diff)
print("To jest 2^-52 =", 2-52)

print("\nWniosek: aby na najmłodszym bicie mantysy pojawiło się 1,")
print("należy dodać do 1.0 wartość 2^-52")

print("\nCzy to najmniejsza liczba > 1?")
print("Tak – jest to najmniejsza liczba dokładnie reprezentowana większa od 1")

print("\nNajwiększa liczba z wykładnikiem 0")
max_val = 2 - 2**-52
print("Wartość:", max_val)
print("hex:", max_val.hex())

print("\nWniosek:")
print("- mantysa składa się z samych jedynek")
print("- jest to największa liczba w przedziale [1,2)")

print("\nPodsumowanie:")
print("- w przedziale [1,2) jest dokładnie 2^52 liczb reprezentowalnych")
print("- najmniejszy krok (ULP) przy 1 wynosi 2^-52")
print("- największa liczba ma mantysę złożoną z samych jedynek")
```

Zadanie 2d


```python

wykladniki = [0, 1, 2, 5, -1, -3]


bity_mantysy = 52

print(f"{'Cecha (E)':<10} | {'Przedział':<18} | {'Odległość sąsiadów':<20} | {'Max błąd bezwzględny':<22} | {'Max błąd względny'}")
print("-" * 100)

for E in wykladniki:
    
    poczatek_przedzialu = 2**E
    koniec_przedzialu = 2**(E+1)
    przedzial = f"[{poczatek_przedzialu}, {koniec_przedzialu})"
    
    
    odleglosc = 2**(E - bity_mantysy)
    
    
    max_blad_bezwzgl = 2**(E - bity_mantysy - 1)
    
    
    max_blad_wzgledny = 2**(-bity_mantysy - 1)
    
    
    print(f"{E:<10} | {przedzial:<18} | {odleglosc:<20.4e} | {max_blad_bezwzgl:<22.4e} | {max_blad_wzgledny:.4e}")

print("\nPodsumowanie:")
print(f"W każdym z tych przedziałów znajduje się dokładnie 2^{bity_mantysy} = {2**bity_mantysy} dokładnie reprezentowanych liczb.")
```

Zadanie 6


```python
def newton(f, f_prim, x_0, tol=1e-15):
    x_prev = x_0
    while True:
        fx = f(x_prev)
        fpx = f_prim(x_prev)
        
        if fpx == 0: return None 
        
        x_n = x_prev - (fx / fpx)
        
       
        if abs(x_n - x_prev) < tol or x_n == x_prev:
            return x_n
        
        x_prev = x_n

def f(x):
    return math.cos(x) - x

def f_prim(x):
    return -math.sin(x) - 1

root = newton(f, f_prim, 0.5)
print(f"Znaleziony pierwiastek: {root}")
```
